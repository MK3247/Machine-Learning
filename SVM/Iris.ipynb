## Iris Classification

```python
import pandas as pd 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target 
df.head(50)
```

```python
x = df[df.target == 0]
xx = df[df.target == 1]
xxx = df[df.target == 2]
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
df0
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'red', marker = 'd')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'green', marker = '+')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Using Sepal Length & Width to differentiate flowers')
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'red', marker = '*')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'green', marker = '+')
plt.xlabel('petal Length')
plt.ylabel('petal Width')
plt.title('Using Petal Length & Width to differentiate flowers')
plt.show()
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower_name'], axis = 1)
y = df['target']
```

```python
y
```

```python
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
X_train
```

## Creating SVM Model 

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[6.4, 2.8, 5.6, 2.2]])
```
