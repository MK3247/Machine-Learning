## Income

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('income.txt')
df
```

```python
plt.scatter(df.Age, df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
```

```python
km = KMeans(n_clusters=3)
```

```python
y_predict = km.fit_predict(df[['Age', 'Income($)']])
y_predict
```

```python
df['Cluster'] = y_predict
df
```

```python
df0 = df[df.Cluster == 0]
df1 = df[df.Cluster == 1]
df2 = df[df.Cluster == 2]
```

```python
df0
```

```python
plt.scatter(df0.Age, df0['Income($)'], color = 'r')
plt.scatter(df1.Age, df1['Income($)'], color = 'g')
plt.scatter(df2.Age, df2['Income($)'], color = 'b')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
scaler  = MinMaxScaler()
scaler.fit(df[['Income($)']])
```

```python
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
xscaler = MinMaxScaler()
xscaler.fit(df[['Age']])
```

```python
df['Age'] = xscaler.transform(df[['Age']])
df
```

```python
plt.scatter(df0.Age, df0['Income($)'], color = 'r')
plt.scatter(df1.Age, df1['Income($)'], color = 'g')
plt.scatter(df2.Age, df2['Income($)'], color = 'b')
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_pred = km.fit_predict(df[['Age', 'Income($)']])
y_pred
```

```python
df.Cluster = y_pred
```

```python
df
```

```python
df0 = df[df.Cluster == 0]
df1 = df[df.Cluster == 1]
df2 = df[df.Cluster == 2]

plt.scatter(df0.Age, df0['Income($)'])
plt.scatter(df1.Age, df1['Income($)'])
plt.scatter(df2.Age, df2['Income($)'])

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'r', marker = 'd')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

## Finding the elbow

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range, sse)
```
