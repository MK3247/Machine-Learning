```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bitcoin.csv')
df.head(30)
```

```python
df.drop('Date',axis = 1, inplace = True)
```

```python
predictionDays = 30
```

```python
df['Prediction'] = df.Price.shift(-predictionDays)
```

```python
df
```

```python
X = np.array(df.drop('Prediction', axis = 1))
len(X)
```

```python
X = X[:len(X) - predictionDays]
len(X)
```

```python
Y = np.array(df['Prediction'])
Y = Y[:-predictionDays]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
pred_days = np.array(df.drop('Prediction', axis = 1))[-predictionDays:]
print(pred_days)
```

```python
from sklearn.svm import SVR
```

```python
model = SVR(kernel = 'rbf', gamma = 0.00001, C=1e3)
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
print("Prediction for the next 30 days")
print()
prediction = model.predict(pred_days)
for i in prediction:
    print(i)
```
