## Salaries - Over 100K

```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.txt')
```

```python
df
```

```python
inputs = df[['company', 'job', 'degree']]
```

```python
output = df[['salary_more_then_100k']]
```

```python
inputs
```

```python
output
```

### Since ML algorithms work best with numbers...

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_name'] = le_company.fit_transform(inputs['company'])
inputs['job_name'] = le_company.fit_transform(inputs['job'])
inputs['degree_name'] = le_company.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs.drop(['job', 'degree'], axis = 1, inplace = True)
```

```python
inputs
```

## Creating Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(inputs, output)
```

```python
model.predict([[1,1,1]])
```

```python
output
```

```python
model.score(inputs, output)
```

The above model gives 100% accuracy because the same dataset was used for training and testing 

What happens when we split the dataset? Let's see what happens

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

The accuracy drops from 100% to 75% when the dataset is split and the model is tested using the test dataset
