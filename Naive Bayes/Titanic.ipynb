```python
import pandas as pd 

df = pd.read_csv('titanic.txt')
df.head(20)
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True)
df.head(20)
```

```python
xx = df.drop('Survived', axis = 1)
xx
```

```python
yy = df.Survived
yy
```

```python
dummies = pd.get_dummies(xx.Sex)
dummies
```

```python
df= pd.concat([xx, dummies], axis = 1)
df.drop('Sex', axis = 1, inplace = True)
df.head(20)
```

```python
x = xx['Age'].mode()
```

```python
df.Age.fillna(24.0, inplace = True)
```

```python
df
```

```python
df.info()
```

### Splitting Data Set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, yy, test_size = 0.2)
```

```python
x_train
```

### Creating Naive Bayes Model 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test[0:10])
```

```python
model.predict_proba(x_test[:10])
```
