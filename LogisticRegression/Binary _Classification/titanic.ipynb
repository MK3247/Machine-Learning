## Creating a sub dataset

```python
subdf = df[['Pclass', 'Pclass', 'Age', 'SibSp', 'Sex', 'Embarked']]
```

```python
subdf
```

## Dummy Variables

Creating dummy variables for:

* Sex 
* Embarked

```python
sex_dummies = pd.get_dummies(df.Sex, prefix = 'Sex')
```

```python
df_with_sex_dummies = pd.concat([subdf, sex_dummies], axis = 1)
```

```python
dataset = df_with_sex_dummies
```

```python
dataset
```

```python
dataset.drop('Sex', axis = 1, inplace = True)
```

```python
dataset
```

```python
embarked_dummies = pd.get_dummies(df.Embarked, prefix = 'Embarked')
df_embarked_dummies = pd.concat([dataset, embarked_dummies], axis = 1)
df_embarked_dummies
```

```python
xx = df_embarked_dummies
```

```python
xx.drop('Embarked', axis = 1, inplace = True)
```

```python
xx
```

```python
x = xx
x
```

```python
y = df.Survived
y
```

## Creating Model for Prediction

Spliting Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
X_train
```

Creating Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test, y_test)
```

### Saving Model

```python
import joblib
```
